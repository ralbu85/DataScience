# 그래프 그리기(Matplotlib)

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Plot 

```python
x=np.arange(200)
y=x**2
```

```python
plt.plot(x,y)
```

## Line Plot with axes and title

```python
x=np.arange(200)
y=x**2
```

```python
plt.plot(x,y)
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('x vs y')
```

## Line Plot with multiple lines

```python
x=np.arange(200)
y=x**2
new_y=x**2-100*x
```

```python
plt.plot(x,y)
plt.plot(x,new_y)
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('x vs y')
plt.legend(['y','new_y'])
```

## Scatter Plot

```python
x=np.array([3,4,7,10,5,2])
y=np.array([2,3,6,8,4,3])
plt.scatter(x,y)
```
